# "with" Block

```python
## no need to close ##

with open("./mahpara.txt") as file:
    y = file.readlines()
    print(y)
```

### Applying File Modes

```python
with open("./mahpara.txt", mode = 'r') as file:   # r = read only
    y = file.readlines()
    print(y)
```

```python
# w = write

with open("./mahpara.txt", mode ='w') as file:
    y = file.readlines()           # don't use read or readlines for 'w'
    print(y)
```

```python
# w = write

with open("./mahpara.txt", mode = 'r') as file:
    file.write("aaaaaaaaaaaaaaaaaaaaaaaaaaaaa")
```

```python
# r+ = read and write, w+ = write and read

with open("./mahpara.txt", mode ='r+') as file:
    y = file.readlines()           
    print("before: ====== ", y)
```

```python
# r+ = read and write, w+ = write and read

with open("./mahpara.txt", mode ='r+') as file:
    y = file.readlines()           
    print("before: ====== ", y)
    file.write("\nPakistan zinda bad")
    y = file.readlines()
    print("after: ====== ", y)
```

```python
# r+ = read and write, w+ = write and read

with open("./mahpara.txt", mode ='r+') as file:
    y = file.readlines()           
    print("before: ====== ", y)
    file.write("\nNED University")
    file.seek(0)           # to move the cursor again to zero position/then it will read from starting
    y = file.readlines()
    print("after: ====== ", y)
```

```python
# r+ = read and write, w+ = write and read

with open("./mahpara.txt", mode ='r+') as file:
    y = file.readlines()           
    print("before: ====== ", y)
    file.write("\nHello World")
    file.seek(0)           # to move the cursor again to zero position/then it will read from starting
    y = file.readlines()
    print("after: ====== ", y)
```

# Append Mode

```python
with open("xyz.txt", "a+") as f:
    f.write("A")
    
```

```python
with open("xyz.txt", "a+") as f:
    f.write("A")
    print(f.read())
```

```python
with open("xyz.txt", "a+") as f:  # a = append, always adds in the last
    f.write("A")
    f.seek(0)
    print(f.read())
```

```python
with open("xyz.txt", "a+") as f:  # a = append, a+ = append and read, always adds in the last
    f.write("\nhello world")
    f.seek(0)
    print(f.read())
```

```python
# only using "w" will overwrite the whole file

with open("xyz.txt", "w") as f: 
    f.write("hello world")
```
