# LOOP

* while
* for 
* controls (optional)
    * break
    * continue
    * pass
    
**Three things are important**
  * counter
  * logic
  * increment/decrement

````python
counter = 0  # counter

while counter < 10:   # logic
    print(counter)
    counter += 1    # increment
````

````python
counter = 0  # counter

while counter < 10:   # logic
    print(counter)
    counter += 1    # increment
else:
    print("Hello")
````

````python
counter = 100

while counter > 0:
    print(counter)
    counter -= 1
````

### Break

````python
counter = 1

while counter < 1000000:
    print(counter)
    if counter == 5:
        break
    counter +=1    
````

### Continue

````python
i = 0

while i < 9:
    i += 1
    if i == 3:   # skips/ignores 3
        continue
    print(i)
````

````python
i = 0

while i < 9:
    i += 1
    if i in [3, 4, 5]:   # skips/ignores 3 to 5
        continue
    print(i)
````

````python
i = 1
print(i)

i += 1   # add 1 in previous value
print(i)

i += 1   # add 1 in previous value
print(i)
````

````python
#        0    1    2    3
name = ['a', 'b', 'c', 'd']

# print(len(name))   # output 4

i = 0    # counter

while i < len(name):
    print(name[i])
    i += 1
````

````python
#        0    1    2    3
name = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']


i = 0    # counter

while i < len(name):    # dynamic length
    print(name[i])
    i += 1
````

````python
## for reverse order ##

#count   1    2    3    4
#index   0    1    2    3
name = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']


i = len(name)-1    # counter

while i >= 0:    # dynamic length
    print(name[i])
    i -= 1
````

````python
names = ["Tom", "Jerry", "Oggy"]
ages = [20, 22, 25]
courses = ["AI", "DS", "ML"]

for name, age, course in zip(names, ages, courses):   # zip with 'for' loop
    print(f"Hello Mr. {name}. Welcome to our {course} class. You are {age} years old.")
````

````python
## login system ##

users = ['a', 'b', 'c']
passwords = ['1', '2', '3']

ui = input("Please enter username:\t")
pi = input("Please enter password:\t")

for u, p in zip(users, passwords):
    if ui == u and pi == p:
        print(f"Welcome {u}")
        break
else:
    print("Not Valid User")
````
